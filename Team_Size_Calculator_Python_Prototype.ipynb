```
def wrapper_dict_selector(dict_name):
  '''
  This function is created for picking specific entries from the configuration dictionaries through wrapper_manpower_estimator
  This is to enable everyone to understand the functionality using the wrapper function. Doesn't need Javascript translation
  '''
  key_list = [key for key in dict_name.keys()]
  pick_dict= dict(zip([i for i in range(len(key_list))], key_list))
  for key, value in pick_dict.items():
    print('Pick {} for {}'.format(key, value))
  inp = input('One Option from above: ')
  try:
    inp = int(inp)
  except ValueError:
    inp=0
  if int(inp) not in [i for i in range(len(key_list))]:
    inp= 0
  else:
    inp = int(inp)
  val = dict_name[pick_dict[inp]]
  print('\n')
  return(val)
```

```
def wrapper_manpower_estimator():
  '''
  This wrapper function is written in lieu of a UI to accept free form inputs and selection of option to run the manpower_calculation function. This need not be translated to
  javascript. This is to enable everyone understand the functionality.
  '''
  eu = int(input('Number of existing Use Cases [integer value only >=0]: '))
  nu = int(input('Number of  New Use Cases [integer value only > 0]: '))
  da = int(input('Number of Data Analytst/ Data Scientists [integer value only > 0]: '))
  et = int(input('Number of Data Management/ ETL/ELT Data Engineers [integer value only > 0]: '))
  im = int(input('Number of Data Ops/ ML Ops Engineers [integer value only > 0]: '))
  print('Choose Primary Type of Work:')
  allocation = wrapper_dict_selector(base_allocations)
  print('Choose Level of Technical Complexity:')
  v1= wrapper_dict_selector(Tech_Complexity)
  print('Choose Level of Business Complexity:')
  v2= wrapper_dict_selector(Biz_Complexity)
  print('Choose Level of Data Complexity:')
  v3= wrapper_dict_selector(Data_Complexity)
  print('Choose Level of Engagement:')
  v4= wrapper_dict_selector(Engagement)
  print('Choose Level of Business Sponsorship:')
  v5= wrapper_dict_selector(Business_Sponsorship)
  print('Choose Level of Integration/ Automation:')
  v6= wrapper_dict_selector(repeat_pct)
  print('Done!! Here are your custom recommendations:')
  return(eu,nu,da,et,im,allocation,v1,v2,v3,v4,v5,v6)

```

```
def manpower_estimator(eu=1,nu=5,da=1,etl=1,im=1,allocation=[.3,.3,.4],v1=1.4,v2=1.2,v3=1.2,v4=0.9,v5=1.05,v6=.2):
  '''
  --- NEEDS JAVASCRIPT CONVERSION----
  This function estimates manpower for future analytics/ data science (DS) initiatives based on current staffing patterns and heuristics based on industry best practices
  Arguments:
  eu: Current existing analytics/ DS Use Cases or initiatives. Should be a integer input
  nu: New Analytics/DS  Use Cases or initiatives. Should be a integer input > 0
  da: Number of Data Analyst/ Data Scientists across all use cases/ initiatives. Should be a integer input > 0
  etl: Number of Data Management/ ETL/ELT Data ingestion staff across all cases/ initiatives. Should be a integer input > 0
  im: Number of implementation/ DataOps/ MLOps staff across all cases/ initiatives. Should be a integer input > 0
  allocation: A list with proportionate break up of manpower across three roles for primary work type. This is a list of decimals summing to 1, 
              that constitutes the values in the base_allocation dictionary
  v1: This is a numerical value that denotes the level of technical complexity of the Technology environment. This is a value from the Technical_Complexity Dictionary
  v2: This is a numerical value that denotes the level of business complexity of the business model. This is a value from the Business_Complexity Dictionary
  v3: This is a numerical value that denotes the level of data complexity of the upstream/ downstream systems. This is a value from the Data_Complexity Dictionary
  v4: This is a numerical value that denotes level of engagement in the Data teams. Meaning effectiveness of interactions. This is a value from the Engagement Dictionary
  v5: This is a numerical value that denotes the level of sponsorship from Top Management. This is a value from the Business_Sponsorship dictionary
  v6: This is a numerical value that denotes the level of integration/ automation in the data science process that helps in reducing headcount. This is a numerical value from the repeat_pct dictionary
  Output:
  The output is a dictionary with the following key- value pair
  1.Total Staff': Total staff across existing and new use cases. Output as xx.y e.g. 19.0
  10.Annualized People Cost': Annualized cost across all roles Output as xxxxxxxx.y e.g. 1755000.0,
  11.Additional Staff': New Hires required. Output as xx.y e.g. 16.0,
  2.DA/DS Staff': Total Data Analyst/ Data Scientists across existing and new use cases. Output as xx.y e.g. 5.0
  3.ETL Staff': Total Data Ingestion/ ETL/ELT  Engineers across existing and new use cases. Output as xx.y e.g. 5.0
  4.Data/ML Ops Staff': Total Data Ops/ ML Ops Engineers across existing and new use cases. Output as xx.y e.g. 7.0
  5.Managers': Number of Managers Output as xx.y e.g. 1.0
  6.Directors': Number of Director Output as xx.y e.g. 0.0
  7.VPs': Number of Vice Presidents Output as xx.y e.g. 0.0
  8.Existing Staffing Rate': Existing Staff/ Existing Use Cases. output as xx.y e.g. 3.0
  9.New Staffing Rate': Total Staff/ (Existing Use Cases+ New Use Cases). output as xx.y e.g. 3.17
  '''
  das_staffing = max(da,1)/max(eu,1)
  etl_staffing = max(etl,1)/max(eu,1)
  imp_staffing = max(im,1)/max(eu,1)
  existing_staff = da+etl+im
  current_staffing= (da+etl+im)/max(eu,1)
  das_new = round(max(v3,v2)*current_staffing*(eu+nu)*allocation[1]*(1-v6)*v5/v4,0)
  etl_new = round(max(v1,v3)*current_staffing*(eu+nu)*allocation[0]*(1-v6)*v5/v4,0)
  imp_new = round(max(v2,v1)*current_staffing*(eu+nu)*allocation[2]*(1-v6)*v5/v4,0)
  etl_leads = etl_new//6
  etl_staff = etl_new - etl_leads
  das_leads = das_new//6
  das_staff = das_new- das_leads
  imp_leads = imp_new//6
  imp_staff = imp_new- imp_leads
  all_staff = etl_staff+etl_leads+das_leads+das_staff+imp_leads+imp_staff
  managers= min(all_staff//15,4)
  directors = min(all_staff//25,1)
  vps = min(all_staff//70,1)
  annual_cost = 85000*all_staff+140000*managers+225000*directors+275000*vps
  new_staff= all_staff- existing_staff
  old_rate = round(existing_staff/max(eu,1),2)
  new_rate = round(all_staff/max((nu+eu),1),2)
  return({'3.ETL Staff': etl_staff,'2.DA/DS Staff':das_staff,'4.Data/ML Ops Staff':imp_staff,\
          '1.Total Staff': all_staff, '8.Existing Staffing Rate':old_rate, '9.New Staffing Rate': new_rate, '5.Managers': managers, \
          '6.Directors': directors, '7.VPs': vps, '10.Annualized People Cost': annual_cost, '11. Additional Staff':new_staff})
```

```
# Weights adjusted to redact proprietary information
base_allocations= {"Descriptive Business Analytics":[.1,.1,.1], \
  "Diagnostic Business Analytics":[.1,.1,.2], \
  "Prescriptive/ Decision Support":[.3,.3,.3], \
  "Automation/AI":[.2,.1,.3]}
Tech_Complexity = dict(zip(['3.Low Complexity','2.Moderate Complexity','1.High Complexity'],[.1,.1,.1]))
Biz_Complexity = dict(zip(['3.Low','2.Moderate','1.High'],[.1,.1,.1]))
Engagement = {'3.Average- 70%':.4, '2.Above Average- 80%':.4, '1.Excellent- 90%!':.4}
Data_Complexity = dict(zip(['3.Low Data Challenges','2.Moderate Data Challenges','1.Heavy Data Challenges'],[.9,.9,.9]))
Business_Sponsorship = dict(zip(['3.Moderate','2.High-Moderate','1.High'],[1,1,1]))
repeat_pct = dict(zip(['4.None','3.Low Integration & Automation', '2.Moderate Integration & Automation', '1.High Integration & Automation'],[0,.1,.20,.40]))
```

```
def main():
  '''
  This is the main function. The enclosed dictionaries are not be changed 
  '''
  print("Welcome to the Analytics Manpower Estimator!!")
  
  eu,nu,da,etl,im,allocation,v1,v2,v3,v4,v5,v6= wrapper_manpower_estimator()
  result_dict= manpower_estimator(eu,nu,da,etl,im,allocation,v1,v2,v3,v4,v5,v6)
  print(result_dict)

if __name__ == "__main__":
    main()

```
